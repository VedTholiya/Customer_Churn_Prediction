```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("Churn_Modelling.csv")
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,3:-1]
y = dataset.iloc[:,-1].values
```

```python
print(X['Geography'].value_counts())
print(X['Gender'].value_counts())
print(X['NumOfProducts'].value_counts())
print(X['HasCrCard'].value_counts())
print(X['IsActiveMember'].value_counts())
```

```python
unique, counts = np.unique(y, return_counts=True)
print(np.asarray((unique, counts)))
```

```python
X=pd.get_dummies(X,drop_first=True)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.15 , stratify=y , random_state=0)
```

```python
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred_l = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
cm_l = confusion_matrix(y_test,y_pred_l)
print(cm_l)
print(accuracy_score(y_test,y_pred_l))
```

```python
print(classification_report(y_test,y_pred_l))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
